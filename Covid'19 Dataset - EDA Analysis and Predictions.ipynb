# Feature transformation

### Transforming categorical data into numerical data

```python
from sklearn.preprocessing import LabelEncoder
e = LabelEncoder()
```

```python
for column in df:
    if column in df.columns:
        df[column] = e.fit_transform(df[column])
df.head(10)
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), cmap = "coolwarm", annot=True)
plt.show()
```

# Feature selection:

### After analysing the correlations above, we can eliminate some rows as they don't add any value to the model

```python
df.columns
```

```python
column_drop = ['Running Nose', 'Asthma', 'Chronic Lung Disease', 'Headache','Heart Disease',
               'Diabetes','Fatigue ','Gastrointestinal ','Wearing Masks','Sanitization from Market']
df = df.drop(columns = column_drop, axis=1 )
df.head(2)
```

# Developing Regression Model

```python
X = df.drop('COVID-19', axis=1)
y = df['COVID-19']
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
y_pred = logmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, f1_score 
```

```python
print('Accuracy of testing data:', accuracy_score(y_test, y_pred)*100,'%')
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
sns.heatmap(cm, cmap="PuBu", annot=True)
```

```python
print('Crossvalidation Score:',cross_val_score(logmodel, X_test, y_test, cv=10).mean())
```

```python
print('f1 score:', f1_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
